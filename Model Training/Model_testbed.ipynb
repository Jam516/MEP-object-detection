```python
def note_max(x,y,ax=None):
    #label max point on plot
    xmax = x[np.argmax(y)]
    ymax = y.max()
    text = "Step={}, AR={:.3f}".format(xmax, ymax)
    if not ax:
        ax=plt.gca()
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    arrowprops = dict(arrowstyle="->", connectionstyle="angle, angleA=0, angleB=300")
    kw = dict(xycoords='data', textcoords='axes fraction',
             arrowprops=arrowprops, bbox=bbox_props, ha='right', va='top')
    ax.annotate(text, xy=(xmax, ymax), xytext=(0.94,0.5), **kw)
```

```python
def pltmAP(df):
    #plot mAP vs steps for a specific test 
    df = df[df['Step']<20000]
    x = df['Step']
    y = df['Value']
    fig, ax = plt.subplots()
    ax.plot(x, y)
    
    ax.set(xlabel='Steps', ylabel='mAP@0.5', title='')
    note_max(x,y)
    ax.set_ylim(0, 1)
    
#     plt.show()
    plt.savefig('Fig55.png')
```

```python
def pltmultiLoss(dflist, nlist):
    #plot loss vs steps for a specific test 
    
    dflist[0] = dflist[0][dflist[0]['Step']<20000]
    dflist[1] = dflist[1][dflist[1]['Step']<20000]
    dflist[2] = dflist[2][dflist[2]['Step']<20000]
    dflist[3] = dflist[3][dflist[3]['Step']<20000]
    
    fig, ax = plt.subplots()
    x = dflist[0]['Step']
    y = dflist[0]['Value']
    ax.plot(x, y, label = nlist[0])
    x1 = dflist[1]['Step']
    y1 = dflist[1]['Value']
    ax.plot(x1, y1, label = nlist[1])
    x2 = dflist[2]['Step']
    y2 = dflist[2]['Value']
    ax.plot(x2, y2, label = nlist[2])
    x3 = dflist[3]['Step']
    y3 = dflist[3]['Value']
    ax.plot(x3, y3, label = nlist[3])
    
    ax.set(xlabel='Steps', ylabel='Loss', title='')
    ax.set_ylim(1, 1.5)
    ax.legend()
    
#     plt.show()
    plt.savefig('Fig55.png')
```

```python
def pltAR(df):
    #plot AR vs steps for a specific test 
    df = df[df['Step']<20000]
    x = df['Step']
    y = df['Value']
    fig, ax = plt.subplots()
    ax.plot(x, y)
    
    ax.set(xlabel='Steps', ylabel='AR@10', title='')
    note_max(x,y)
    ax.set_ylim(0, 1)
    
#     plt.show()
    plt.savefig('Fig55.png')
```

```python
def compare(dflist):
    #return list of mAP values at the same step for multiple tests
    ptlist = []
    for i in dflist:
        x = i['Step']
        y = i['Value']
        ptlist.append(np.interp(20000, x, y))
    
    return ptlist
```

```python
def maxcompare(dflist):
    #return list of highest mAP values for multiple tests
    ptlist = []
    for i in dflist:
        y = i['Value']
#         y = rollingAvg(i)
        ptlist.append(y.max())
    
    return ptlist
```

```python
def pltsensitivity(dflist, xlist, parameter):
    #plot sentivity analysis for a set parameter
    ptlist = maxcompare(dflist)

# To plot in line style
# ---------------------------------------------
    fig, ax = plt.subplots(figsize=(8,6))
    ax.plot(xlist, ptlist, marker='o')
    ax.set(xlabel=parameter, ylabel='AR@10')
    plt.savefig('Fig44.png')
    
# To plot in scatter style
# ----------------------------------------------
#     fig, ax = plt.subplots()
#     ax.set(xlabel=parameter, ylabel='mAP@0.5')
#     plt.scatter(xlist, ptlist)
#     z = np.polyfit(xlist, ptlist, 1)
#     p = np.poly1d(z)
#     plt.plot(xlist,p(xlist),"r--")
#     plt.savefig('Fig41.png')
```

```python
def pltdetail(easy_lst, medium_lst, hard_lst, xlist, parameter):
    #plot sentivity analysis for a set parameter segmented by size ranges
    easy = maxcompare(easy_lst)
    medium = maxcompare(medium_lst)
    hard = maxcompare(hard_lst)

    fig, ax = plt.subplots()
    ax.plot(xlist, easy, marker='o', color='green', label = 'small')
    ax.plot(xlist, medium, marker='o', color='blue', label = 'medium')
    ax.plot(xlist, hard, marker='o', color='red', label = 'large')
    ax.set(xlabel=parameter, ylabel='mAP@0.5')
    ax.legend()
    plt.savefig('Fig46.png')
```

```python
def pltstagnation(df):
    #plot % change in mAP over past 100 steps
    ptlist = []
    difflist = []
    chngelist = []
    x = df['Step']
    y = df['Value']
    for step in range(0, 20100, 100):
        ptlist.append(np.interp(step, x, y))
    for i in range(0, 200):
        difflist.append(ptlist[i+1]-ptlist[i])
    for i in range(0, 200):
        chngelist.append((difflist[i]/ptlist[i])*100)
    
    fig, ax = plt.subplots()
    ax.plot(range(0, 20000, 100), chngelist)
    ax.set(xlabel='Steps', ylabel='% change', title='')
    ax.set_ylim(-1, 10)
    plt.savefig('Fig52.png')
```
